```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = {
    "title": [
        "Inception", "Titanic", "Avengers", "The Dark Knight",
        "Interstellar", "Gladiator", "Joker", "Avatar",
        "Black Panther", "Forrest Gump"
    ],
    "genre": [
        "Sci-Fi", "Romance", "Action", "Action",
        "Sci-Fi", "Drama", "Drama", "Sci-Fi",
        "Action", "Drama"
    ],
    "rating": [8.8, 7.9, 8.0, 9.0, 8.6, 8.5, 8.4, 7.8, 7.3, 8.8]
}

movies = pd.DataFrame(data)
movies
```

```python
# Check dataset info
movies.info()

# Check for missing values
movies.isnull().sum()
```

```python
mean_rating = movies["rating"].mean()
median_rating = movies["rating"].median()
mode_rating = movies["rating"].mode()[0]

print("Mean Rating:", mean_rating)
print("Median Rating:", median_rating)
print("Mode Rating:", mode_rating)
```

```python
plt.figure(figsize=(6,4))
sns.histplot(movies["rating"], bins=5, kde=True)
plt.title("Distribution of Movie Ratings")
plt.xlabel("Rating")
plt.ylabel("Frequency")
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.boxplot(x=movies["rating"])
plt.title("Box Plot of Movie Ratings")
plt.show()
```

```python
top_movies = movies.sort_values(by="rating", ascending=False).head(5)
top_movies
```

The top-rated movies were identified by sorting ratings in descending order

```python
top_genres = movies.groupby("genre")["rating"].mean().sort_values(ascending=False)
top_genres
```

```python
plt.figure(figsize=(10,5))
sns.barplot(data=top_movies, x="rating", y="title")
plt.title("Top 5 Rated Movies")
plt.xlabel("Rating")
plt.ylabel("Movie Title")
plt.show()
```
